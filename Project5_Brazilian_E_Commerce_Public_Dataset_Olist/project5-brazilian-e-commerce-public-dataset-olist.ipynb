# Brazillian E-Commerce Analysis

This project analyzes the Brazillian E-Commerce dataset provided by Olist.
The objective is to explore customer behavior, product performance, delivery efficiency, payment patterns, and seller performance.

Through comprehensive data analysis, the project aims to uncover business insights that can support strategic decision-making and improve overall platform performance.

The analysis covers mutiple aspects of the e-commerce ecosystem, including sales trends, customer purchasing patterns, logistics performance, and marketplace selelr behavior.

## Business Questions
This analysis aims to answer the following key business questions:
1. How has the platform's sales performance evolved over time?
2. What are the characteristics of customer purchasing behavior?
3. Which product categories drive the most sales and revenue?
4. How efficient is the delivery process and how does it affect customer satisfication?
5. What are the most commonly used payment methods and installment behaviors?
6. How do sellers perform in terms of sales, delivery reliability, and customer ratings?

## Dataset Overview
The dataset used in this project comes from the Brazillian e-commerce platfom Olist and contains transactional data from 2016 to 2018.

The dataset includes multiple tables that represent different aspects of the marketplace:

- Orders: order status, timestamps. and delivery information
- Customers: customer identifiers and locations
- Order Items: product details within each order
- Products: product categories and attributes
- Payments: payment methods and installment details
- Reviews: customer review scores
- Sellers: seller identifies and locations
- Geolocation: geographic information for customers and sellers

These tables were merged to create a unified dataset for comprehensive analysis.

## Data Preparation
Before performing the analysis, several preprocessing steps were conducted:

- Imported and loaded multiple datasets
- Checked for missing values and data consistency
- Converted timestamp columns to datetime format
- Created additional time-based features such as order year and month
- Merged datasets to create a comprehensive analytical table
- Calculated additional metrics such as revenue, delivery time, and delivery delay

After preprocessing, the final dataset contains 119,143 records and 39 variables.

## Sales Performance Analysis



The monthly revenue trend shows a clear growth pattern throughout 2017, indicating strong expansion of the e-commerce platform during that period.

Revenue steadily increased from early 2017 and reached its peak between October 2017 and January 2018. This suggests that platform experienced its highest sales activity during the late 2017 period.

After the peak period, revenue levels remained relatively stable before declining in the final months of the dataset. However, the decline is likely influenced by incomplete data for the last recorded months rather than a true decrease in business performance.

Overall, the revenue trend suggests a period of significant platform growth followed by stabilization.



The order volume trend closely follows the revenue pattern, showing a steady increase throughout 2017.

The number of orders gradually increased and reached a peak around late 2017, which aligns with the revenue peak observed during the same period. This indicates that the growth in revenue was primarily driven by an increase in the number of orders rather than changes in pricing.

Similar to the revenue trend, order volume appears to decline in the final months of the dataset. This decline is likely due to incomplete data coverage for the last period rather than a true drop in customer demand.



The Average Order Value (AOV) shows a noticeable fluctuation during the early period of the dataset. A sharp decline occurred between October 2016 and December 2016, followed by a recovery in early 2017.

From 2017 onward, the AOV remained relatively stable, generally staying above 150. This indicates that while order volume was increasing during 2017, the average spending per order remained consistent.

The stability of AOV suggests that the overall revenue growth during 2017 was mainly driven by increasing transaction volume rather than higher spending per order.

## Customer Behavior Analysis



The distribution of orders per customer shows that the vast majority of customers made only a single purchase on the platform.

Approximately 90% of customers placed only one order, while a much smaller portion of customers made repeat purchases. This indicates that repeat purchasing behavior is relatively limited within the dataset.

The repeat purchase rate is approximately 3.12%, suggesting that only a small percentage of customers returned to make additional purchases.



The repeat purchase rate provides additional insight into customer retention behavior. With a repeat purchase rate of approximately 3.12%, the platform appears to rely heavily on new customers rather than repeat buyers.

This pattern suggests that customer acquisition may play a larger role in driving platform activity compared to customer retention.



The distribution of customer revenue is highly skewed, with the majority of customers contributing relatively small amounts of revenue.

The histogram shows a very large concentration of customers at the lower end of the revenue range, while only a small number of customers generate significantly higher revenue.

This long-tail distribution indicates that revenue contribution is unevenly distributed across the customer base, with a small group of customers accounting for a disproportionate share of total revenue.

## Product & Category Analysis



The analysis of product categories by order volume shows that several categories dominate customer purchases.

The categories with the highest number of orders include **bed_bath_table**, **health_beauty**, and **sports_leisure**. These categories represent the most frequently purchased products on the platform.

The high order volume in these categories suggests strong customer demand for household, personal care, and lifestyle-related products.



When analyzing revenue contribution by product category, **health_beauty**, **watches_gifts**, and **bed_bath_table** emerge as the top revenue-generating categories.

While some categories lead in order volume, others contribute more significantly to overall revenue. For example, **watches_gifts** appears among the top revenue categories despite not being among the top categories by order count.

This suggests that certain categories may have higher product prices or larger transaction values.



Customer review analysis shows that some product categories receive lower satisfaction scores compared to others.

Among the lowest-rated categories, **furniture_bedroom** appears with the lowest average review score. Lower ratings may indicate potential issues related to product quality, delivery experience, or customer expectations.

Monitoring categories with lower review scores is important for identifying areas where customer experience could potentially be improved.



In contrast, several categories demonstrate high customer satisfaction based on review scores.

Categories such as **cds_dvds_musicals**, **fashion_children_clothes**, and **books_general_interest** appear among the highest-rated product categories.

These categories consistently receive strong customer ratings, suggesting that customers generally have positive purchasing experiences with these products.

## Delivery & Logistics Performance



The analysis of delivery time shows that the average delivery duration is approximately 12 days from purchase to customer delivery.

The distribution indicates that most deliveries fall within a relatively moderate time range, although there are some orders with significantly longer delivery times. These outliers suggest that certain orders may experience extended delivery durations due to logistical factors.



To evaluate delivery reliability, the difference between actual delivery time and estimated delivery time was calculated.

The analysis shows that approximately **7.11% of orders were delivered later than the estimated delivery date**, indicating that the majority of deliveries were completed within the promised timeframe.

Although the late delivery rate is relatively small compared to the total number of orders, delayed deliveries still represent a notable portion of the customer experience.



Customer review scores were analyzed in relation to delivery performance.

Orders delivered on time have an average review score of approximately **4.13**, while orders that experienced delivery delays have a significantly lower average review score of around **2.44**.

This represents a difference of approximately **1.69 points**, suggesting that delivery delays are strongly associated with lower customer satisfaction ratings.

## Payment Behavior Analysis



The distribution of payment methods shows that **credit card** is the most commonly used payment type on the platform.

Credit card transactions account for the majority of orders, significantly exceeding other payment methods such as debit card, boleto, and voucher payments. This indicates that credit cards are the dominant payment method among customers.



The analysis of installment usage shows that the most common payment option is **a single installment**, which represents one-time payment without installment splitting.

The number of orders using a single installment is significantly higher than those using multiple installments. This suggests that most customers prefer to complete their purchases through full payment rather than spreading payments across multiple installments.



The comparison of Average Order Value (AOV) across payment types shows that **credit card payments have the highest average order value**.

This suggests that customers using credit cards tend to make larger purchases compared to those using other payment methods. The flexibility and convenience of credit card payments may contribute to higher spending behavior.

## Seller Performance Analysis



The analysis of seller order volume shows that a small number of sellers account for a large number of transactions on the platform.

The seller with the highest order volume recorded approximately **1,854 orders**, indicating that certain sellers play a major role in driving transaction activity within the marketplace.

This concentration suggests that marketplace performance may depend significantly on a relatively small group of highly active sellers.



When analyzing seller performance by revenue, the results show that the seller generating the highest revenue recorded approximately **258,882 in total sales**.

Interestingly, the seller with the highest revenue is not the same seller with the highest number of orders. This suggests that some sellers may generate higher revenue through higher-priced products rather than larger order volumes.



Seller delivery performance was also analyzed by calculating the average delivery delay for each seller.

The results show that some sellers experience significantly higher delivery delays compared to others. In the dataset, the seller with the highest average delay recorded approximately **167 days**, which represents a substantial deviation from the overall delivery average.

Such extreme delays may indicate logistical challenges, operational inefficiencies, or potential data anomalies that could impact customer experience.

## Key Business Insights

### 1. Extremely Low Customer Retention

The analysis reveals that the repeat purchase rate is only **3.12%**, indicating that the vast majority of customers make only a single purchase on the platform.

This rate is significantly lower than typical e-commerce benchmarks, where repeat purchase rates commonly range between **20% and 40%**. The data also shows that approximately **90% of customers purchase only once**, suggesting that customer retention is extremely limited.

This pattern implies that the platform relies heavily on acquiring new customers rather than maintaining long-term relationships with existing customers. Such a model can lead to higher customer acquisition costs and reduced long-term profitability.

### 2. Delivery Performance Strongly Affects Customer Satisfaction

The analysis indicates a strong relationship between delivery performance and customer review scores.

Orders delivered on time receive an average review score of approximately **4.13**, while orders delivered late receive a significantly lower average score of around **2.44**.

This substantial difference suggests that delivery delays have a major negative impact on customer satisfaction. Maintaining reliable delivery performance is therefore critical for preserving positive customer experiences and protecting platform reputation.

### 3. Revenue Concentration Among a Small Portion of Customers

Customer revenue distribution is highly skewed, with a small number of customers generating a disproportionate share of total revenue.

The revenue histogram shows a strong concentration in the lower spending segments, while only a small group of customers contributes significantly higher order values.

This pattern is consistent with the **Pareto principle**, where a minority of customers generate a large portion of business revenue. Identifying and retaining these high-value customers may provide substantial opportunities for revenue growth.

### 4. Marketplace Activity Concentrated Among a Small Number of Sellers

Seller activity on the platform is also unevenly distributed.

The top seller recorded approximately **1,854 orders**, indicating that a relatively small group of sellers contributes a large share of marketplace transactions.

This concentration suggests that platform performance may depend heavily on a limited number of high-performing sellers. Such dependency may create operational risks if these sellers reduce activity or leave the platform.

## Business Recommendations

### 1. Develop a Customer Membership and Loyalty Program

Given the extremely low repeat purchase rate of 3.12%, improving customer retention should be the platform’s highest strategic priority.

A membership-based loyalty program could encourage customers to return to the platform for future purchases. For example, the platform could introduce reward points, exclusive member discounts, or tier-based benefits that increase as customers make more purchases.

Such programs help strengthen customer relationships and incentivize repeat purchasing behavior, ultimately increasing customer lifetime value.

### 2. Improve Delivery and Logistics Reliability

The analysis shows that late deliveries significantly reduce customer review scores. Orders delivered on time receive an average review score of approximately 4.13, while late deliveries receive an average score of only 2.44.

This indicates that delivery performance plays a critical role in shaping customer satisfaction.

Improving logistics coordination, monitoring seller shipping performance, and optimizing delivery operations could help reduce delays and improve overall customer experience.

### 3. Develop Targeted Strategies for High-Value Customers

Customer revenue distribution is highly skewed, with a small percentage of customers generating a disproportionate amount of revenue.

Identifying these high-value customers and providing personalized offers, exclusive deals, or premium membership benefits could help strengthen their loyalty and increase their lifetime value.

Targeted marketing strategies can therefore maximize revenue by focusing on the most valuable customer segments.

### 4. Strengthen Seller Performance Management

Marketplace activity is concentrated among a relatively small group of sellers. While these sellers contribute significantly to transaction volume, heavy dependence on a limited number of sellers may create operational risks.

The platform could implement seller performance monitoring systems, encourage more sellers to scale their operations, and provide support programs to help smaller sellers grow.

Diversifying seller performance across the marketplace can help create a more stable and resilient platform ecosystem.

## Conslusion

This project analyzed the Brazilian E-Commerce dataset to evaluate marketplace growth, customer behavior, operational performance, and seller dynamics.

The analysis revealed several key findings. First, customer retention is extremely low, with a repeat purchase rate of only 3.12%. The majority of customers make only a single purchase, indicating that the platform relies heavily on acquiring new customers rather than maintaining long-term customer relationships.

Second, delivery performance has a strong influence on customer satisfaction. Orders delivered on time receive significantly higher review scores than those delivered late, highlighting the importance of reliable logistics in maintaining a positive customer experience.

Third, revenue distribution is highly concentrated among a small portion of customers, suggesting that high-value customers represent an important opportunity for targeted retention strategies.

Finally, marketplace activity is unevenly distributed among sellers, with a relatively small number of sellers contributing a large share of total orders.

Based on these findings, the platform should prioritize improving customer retention through membership and loyalty programs, strengthening logistics performance, developing targeted strategies for high-value customers, and supporting a more balanced seller ecosystem.

Overall, these strategies can help improve long-term customer engagement, enhance operational performance, and support sustainable marketplace growth.

```python
#convert date columns
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

print(orders.info())
```

```python
#create time features
orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
orders['order_day'] = orders['order_purchase_timestamp'].dt.day

print(orders.info())
```

```python
#merge core tables
#orders + customer
df = orders.merge(customers, on='customer_id', how='left')
#+order_items
df = df.merge(order_items, on='order_id', how='left')
#+products
df = df.merge(products, on='product_id', how='left')
#+payments
df = df.merge(payments, on='order_id',how='left')
#+reviews
df = df.merge(reviews, on='order_id',how='left')
```

```python
#Code analysis - Sales Performance Analysis
df['revenue'] = df['price'] + df['freight_value']
monthly_revenue = df.groupby(['order_year','order_month'])['revenue'].sum().reset_index()
monthly_revenue.head()
```

```python
#date time
monthly_revenue['date'] = pd.to_datetime(
    monthly_revenue['order_year'].astype(str) + '-' +
    monthly_revenue['order_month'].astype(str)
)
```

```python
#revenue trend visualization

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_revenue, x='date', y='revenue')

plt.title('Monthly Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue')

plt.show()
```
