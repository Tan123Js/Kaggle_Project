## Cohort Retention Heatmap

Purpose: Examine customer repeat purchase behavior over time.



The cohort heatmap shows customer retention over time, measured by momths since first purchase (cohort Index).

**Key observations:**

- Cohort Index 0 (first month) always shows 100% retention, representing the initial purchase.
- From 2010-03 to 2010-08, the cohorts display a "horizontal convex shape" in months 1–5:
    - Retention drops sharply after the first month to ~70–40%
    - A small peak appears mid-period, followed by a decline
    - Maximum cohort index observed is 5, indicating most repeat purchases occur within the first 5 months
- This suggests that customer repeat purchases are concentrated in the first few months, highlighting a critical window for targeted promotions or loyalty programs to increase repeat purchase rates.

## Customer Behavior & Revenue Concentration Analysis

The combined analysis of customer retention (Cohort Heatmap) and revenue distribution (Pareto Analysis) provides a comprehensive view of customer behavior and its impact on business revenue.

**a. Cohort Retention Analysis**:

Cohort Index 0, representing the first month of purchase, consistently shows 100% retention, as all customers are new.

Between 2010-03 and 2010-08, cohorts exhibit a “horizontal convex shape” across Cohort Index 1–5:

- Retention sharply declines after the first month, typically to 70–40%
- A mid-period uptick appears, followed by a subsequent decline
- The maximum observed Cohort Index is 5, indicating that most repeat purchases occur within the first five months

**Implication**: Customer repeat purchases are concentrated in the initial months. This period represents a critical window for targeted promotions, loyalty programs, or engagement campaigns to maximize repeat purchase rates and customer lifetime value.

**b. Customer Revenue Distribution (Pareto Analysis)**:

Approximately 27% of customers generate 80% of total revenue, demonstrating a clear concentration of revenue among a small group of high-value customers.

**Implication**: The business relies heavily on these top customers. Retaining this segment is essential to maintain overall revenue stability. Strategic initiatives could include VIP programs, personalized promotions, or exclusive offers to enhance loyalty and encourage further high-value transactions.

**Integrated Insights**:

The majority of repeat purchases happen early in the customer lifecycle, highlighting the importance of early engagement.
Revenue is highly concentrated among a relatively small group of customers, emphasizing the critical need to retain high-value clients.
Together, these analyses suggest a two-pronged strategic approach:

* Short-term: Design promotions and engagement campaigns targeting new customers within the first 5 months to increase retention.

* Long-term: Implement loyalty and VIP programs focused on top revenue-generating customers to secure sustained business growth.

**Conclusion**:

By combining cohort and Pareto analyses, the business gains actionable insight into when and which customers to focus on, enabling more effective resource allocation, marketing campaigns, and revenue maximization strategies.

## Business Insights

Several important insights were identified from the analysis:

1. Revenue Concentration Risk: Revenue is heavily concentrated in the UK (~84%). Any disruption in this market could severely impact overall revenue.

2. Product Concentration: A few products drive most sales, with “White Hanging Heart T-Light Holder” being the top contributor.

3. Low Average Order Value: Most purchases are small, indicating potential to increase revenue per transaction through bundling or upselling.

4. Customer Retention Declines Quickly: Many customers do not return after the first month; retention stabilizes but remains low.

5. Revenue Dependent on High-Value Customers: ~27% of customers contribute 80% of revenue, highlighting the importance of retaining top customers.

### Business Recommendation

To reduce market dependency and increase customer retention, the company could consider:

1. Market Diversification: Reduce reliance on UK market by expanding into other countries.

2. Increase Average Order Value: Encourage larger purchases via bundling, upselling, or free shipping thresholds.

3. Improve Customer Retention: Implement targeted promotions or loyalty programs during the first 3–6 months after initial purchase to encourage repeat buying.

4. Focus on High-Value Customers: Identify top 20–30% of customers contributing to majority revenue and offer VIP programs or personalized promotions.

## Conclusion
The company is highly dependent on the UK market for revenue. Sales are concentrated among a limited number of decorative products. Customer retention declines significantly after the first purchase month. Targeted post-purchase campaigns may improve long-term customer value.

### Phase 3 User Bahavier Analysis
#### RFM Analyse
To figure out high value customers
* Recency : The days that after bought
* Frequency : Total order times
* Monetary : Total revenue

```python
#RFM calculation
import pandas as pd
import datetime

analysis_date = df_clean['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df_clean.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (analysis_date - x.max()).days, #Recency
    'Invoice':'nunique', #Frequency
    'Revenue': 'sum' #Monetary
})

rfm.rename(columns={'InvoiceDate': 'Recency', 'Invoice': 'Frequency', 'Revenue': 'Monetary'}, inplace=True)
rfm.head()
```

### RFM Scoring
Four level based calculation, to identify customer group

```python
#view the frequency problem
rfm['Frequency'].describe()
#because

#rfm['F_Score'] = pd.qcut(rfm['Frequency'],4,labels=[4,3,2,1]) 
#-> cause “ValueError: Bin edges must be unique: Index([1.0, 1.0, 3.0, 6.0, 270.0], 
#dtype='float64', name='Frequency'). You can drop duplicate edges 
#by setting the 'duplicates' kwarg”

#Then if
#rfm['F_Score'] = pd.qcut(rfm['Frequency'],4,labels=[4,3,2,1],duplicates = 'drop')
#-> cause “ValueError: Bin labels must be one fewer than the number of bin edges“

#So Final:
#rfm['F_Score'] = pd.qcut(rfm['Frequency'],q= 4,labels=False,duplicates = 'drop')+1

#As the result shows, this is a long tail distribution = right side growth
# a common solution is rule-based segmentation (if else)
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'],q= 4,labels=False,duplicates = 'drop')+1
rfm['M_Score'] = pd.qcut(rfm['Monetary'],4,labels=[4,3,2,1])

rfm['RFM_Score'] = rfm['R_Score'].astype(str)+rfm['F_Score'].astype(str)+ rfm['M_Score'].astype(str)
rfm.head()
```

```python
#calculate cohort index（which month）
df_clean = df_clean.dropna(subset=['InvoiceDate'])
df_clean['CohortMonth'] = df_clean.groupby('Customer ID')['InvoiceDate'].transform('min').dt.to_period('M')
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M')

df_clean['CohortIndex'] = (df_clean['InvoiceMonth'] - df_clean['CohortMonth']).apply(lambda x:x.n)

cohort_data = df_clean.groupby(['CohortMonth', 'CohortIndex'])['Customer ID'].nunique().reset_index()

cohort_counts = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='Customer ID')

cohort_sizes = cohort_counts.iloc[:,0]

retention = cohort_counts.divide(cohort_sizes, axis=0)

retention*100
```

```python
#Pareto Analysis（80/20 Rule）

customer_revenue = df_clean.groupby('Customer ID')['Revenue'].sum().reset_index()
customer_revenue = customer_revenue.sort_values(by='Revenue', ascending=False)
print(customer_revenue.head())
```

```python
#cumulative revenue
customer_revenue['CumulativeRevenue'] = customer_revenue['Revenue'].cumsum()
customer_revenue['RevenuePercent'] = customer_revenue['CumulativeRevenue'] / customer_revenue['Revenue'].sum()
print("\nRevenuePercent\n",customer_revenue['RevenuePercent'],"\n\nCumulativeRevenue\n", customer_revenue['CumulativeRevenue'])
```

```python
customer_revenue['CustomerPercent'] = range(1, len(customer_revenue)+1)
customer_revenue['CustomerPercent'] = customer_revenue['CustomerPercent'] / len(customer_revenue)
print("\nCustomerPercent\n",customer_revenue['CustomerPercent'])
```

```python
#pareto chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.plot(customer_revenue['CustomerPercent'], customer_revenue['RevenuePercent'])

plt.xlabel("Customer Percentage")
plt.ylabel("Revenue Percentage")
plt.title("Customer Revenue Distribution (Pareto Analysis)")

plt.show()
```
