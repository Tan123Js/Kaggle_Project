The **Exited** variable is the key indicator used to identify churn behavior and build the predictive model.

## Model Development

After completing the exploratory data analysis, the next step is to build a predictive model to estimate the probability of customer churn.

Logistic Regression was selected as the primary model for this analysis. This algorithm is widely used for binary classification problems where the objective is to predict one of two outcomes. In this case, the model predicts whether a customer will churn (Exited = 1) or remain with the bank (Exited = 0).

Before training the model, several preprocessing steps were performed:

- Irrelevant identifiers such as `RowNumber`, `CustomerId`, and `Surname` were removed.
- Categorical variables were encoded to convert them into numerical representations.
- The dataset was separated into input features (X) and the target variable (y).
- The data was split into training and testing sets to evaluate model performance on unseen data.

The logistic regression model was then trained using the training dataset to learn patterns associated with customer churn.

## Model Evaluation

To assess the effectiveness of the predictive model, several evaluation metrics were used. These metrics help determine how well the model can identify customers who are likely to churn.

The evaluation process includes:

- Accuracy Score
- Confusion Matrix
- ROC Curve and AUC Score

These metrics provide different perspectives on model performance and allow us to better understand the strengths and limitations of the predictive model.

### Model Accuracy

The accuracy score measures the proportion of correct predictions made by the model with 81.13%.

Accuracy is calculated as the number of correct predictions divided by the total number of predictions.

While accuracy provides a general indication of model performance, it is not always sufficient for churn prediction problems because the dataset may be imbalanced. Therefore, additional evaluation metrics such as the confusion matrix and ROC curve are also examined.

### Confusion Matrix

The confusion matrix provides a detailed breakdown of the model's predictions by comparing predicted values with actual outcomes.

It categorizes predictions into four groups:

- True Positives (correctly predicted churned customers)
- True Negatives (correctly predicted retained customers)
- False Positives (customers predicted to churn but actually stayed)
- False Negatives (customers predicted to stay but actually churned)

For churn prediction, false negatives are particularly important because they represent customers who left the bank but were not identified as high risk by the model.



### ROC Curve and AUC Score

The Receiver Operating Characteristic (ROC) curve evaluates the model’s ability to distinguish between churned and retained customers across different probability thresholds.

The Area Under the Curve (AUC) provides a single performance metric summarizing the model’s classification capability.

An AUC score closer to 1 indicates better predictive performance, while a score around 0.5 suggests that the model performs no better than random guessing.

In churn prediction tasks, the ROC-AUC metric is commonly used because it provides a more robust evaluation compared to accuracy alone.

### Feature Engineering

```python
df_model = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
#df_model.columns
print(df_model.columns)
print(df_model.info())
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()
df_model['Gender'] = le.fit_transform(df_model['Gender'])
df_model = pd.get_dummies(df_model, columns=['Geography'], drop_first=True)

y= df_model['Exited']
X = df_model.drop('Exited',axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=1000, random_state =42)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)
#accuracy_score(y_test, y_pred)
```

```python
#accuracy calculation
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy:.2%}")
```

```python
#confusion matrix
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True, fmt='d',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
#ROC & AUC
y_prob = model.predict_proba(X_test_scaled)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test,y_prob)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label = f"AUC = {auc_score:.2f}")
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Feature Importance
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient',key=abs,ascending=False)

print(feature_importance)
```

```python
#risk segmentation
risk_prob = model.predict_proba(X_test_scaled)[:,1]

risk_segment = pd.cut(
    risk_prob,
    bins=[0,0.3,0.6,1],
    labels=['Low Risk','Medium Risk','High Risk']
)
```

```python
risk_df = pd.DataFrame({
    'Churn_Probability': risk_prob,
    'Risk_Level': risk_segment,
    'Actual_Churn': y_test
})

risk_df['Risk_Level'].value_counts()
```

```python
print(risk_df.groupby('Risk_Level', observed=False)['Actual_Churn'].mean())
```

```python
sns.countplot(x='Risk_Level', data=risk_df)
plt.title('Customer Risk Segmentation')
plt.show()
```
